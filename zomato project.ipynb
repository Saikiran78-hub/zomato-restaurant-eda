# Zomato data analysis project

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("project zomato/Zomato data .csv")
print(dataframe)
```

## convert the data type of column - rate

```python
def handleRate(value):
    value=  str(value).split('/')
    value = value[0]
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

## Type of restaurant 

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'], color = "g")
plt.xlabel("type of restaurant")
```

# conclusion - majority of the restaurant falls in dining category

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby("listed_in(type)")["votes"].sum()
result = pd.DataFrame({"votes": grouped_data})
plt.plot(result, c = "m", marker ="o" )
plt.xlabel("Type of restaurant", c = "red", size = 20)
plt.ylabel("votes", c= "r", size = 20)
```

## conclusion - dining restaurants has recieved maximum votes

## 4. Average order spending by couples

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data,color = "c")
```

## conclusion - the majority of couples prefer restaurants with an approximate cost of 300 rupees

## 5. which mode receives maximum rating 

```python
plt.figure(figsize = (6,6))
sns.boxplot(x= "online_order", y = "rate", data = dataframe)
```

## conclusion - offline order received lower rating in comaparison to online order

```python
pivot_table = dataframe.pivot_table(index="listed_in(type)", columns= 'online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot =True, cmap="YlGnBu")
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

## CONCLUSION : Dining restauarants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefered orders in person at restauarants, but prefer online ordering at cafes.
